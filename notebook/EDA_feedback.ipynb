### 3.7 Exploring Data

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Unique values for column '{column}':")
    print(unique_values)
    print()
```

```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
rating_columns=['inflight_wifi_service','departure_arrival_time_convenient','ease_of_online_booking','gate_location','food_and_drink','online_boarding','seat_comfort','inflight_entertainment','on_board_service','leg_room_service','baggage_handling','checkin_service','inflight_service','cleanliness']
for column in rating_columns:
    max_rating = df[df[column] == 5]['id'].count()
    print(f'Number of Passengers rated 5 for {column} : {max_rating}')
print()
for column in rating_columns:
    min_rating = df[df[column] == 0]['id'].count()
    print(f'Number of Passengers rated 0 for {column} : {min_rating}')
```

#### Insight
- baggage_handling service has maximun 5 star rating from passengers
- inflight_wifi_service service has minimum 5 star rating from passengers
- departure_arrival_time_convenient service has maximun 0 star rating from passengers
- baggage_handling has No 0 star rating from passengers
